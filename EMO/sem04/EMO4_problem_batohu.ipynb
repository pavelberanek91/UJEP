## Cvičení 3 - Problém batohu

```python
M = 30 # kapacita batohu
a = [#(kg, cena)
    (1, 100),
    (1, 100),
    (1, 100),
    (1, 100),
    (1, 100),
    (2, 180), 
    (2, 180), 
    (2, 180), 
    (2, 180), 
    (5, 400), 
    (5, 400), 
    (5, 400), 
    (10, 800), 
    (10, 800)]
```

```python
def hodnota_batohu(alpha, a, M):
    cost = 0
    weight = 0
    for i in range(len(alpha)):
        if alpha[i] == "1":
            if (weight + a[i][0]) <= M:
                weight += a[i][0]
                cost += a[i][1]
            else:
                cost -= a[i][1]
                weight += a[i][0]
    return cost

hodnota_batohu("01111", a, M)
```

```python
import random

def random_alpha(size):
    return "".join([str(random.randint(0, 1)) for i in range(size)])

print(random_alpha(len(a)))
```

```python
def gray2dec(alfa):
    dual = [alfa[0]]
    for nextbinary in alfa[1:]:
        dual.append(str(int(dual[-1]) ^ int(nextbinary)))
    return int("".join(dual), 2)

print(gray2dec("1011"))
```

```python
def mutation(alfa, pmut):
    child = "" 
    for i in range(len(alfa)):
        child += str(1 + int(alfa[i]) * -1) if random.random() < pmut else alfa[i]
    return child

print(mutation("1001", pmut=0.25))
```

```python
def random_search(f, tmax, k, n):
    f_max = 0
    alpha_max = None
    for time in range(tmax):
        alpha = random_alpha(size=k*n)
        f_x = f(alpha)
        if f_x > f_max:
            f_max =  f_x
            alpha_max = alpha
        print(f"t: {time}, alpha: {alpha}, f: {f_x}, alpha_max: {alpha_max}, f_max: {f_max}")
    return (alpha_max, f_max)
```

```python
def hill_climb(f, tmax, k, n, c0, pmut):
    f_max = 0
    alpha_max = None
    alpha = random_alpha(size=k*n)
    for time in range(tmax):
        U = [mutation(alpha, pmut) for i in range(c0)]
        alpha = max(U, key=lambda alpha: f(alpha))
        f_x = f(alpha)
        if f_x > f_max:
            f_max =  f_x
            alpha_max = alpha
        print(f"t: {time}, alpha: {alpha}, f: {f_x}, alpha_max: {alpha_max}, f_max: {f_max}")
    return (alpha_max, f_max)
```

```python
def tabu_search(f, tmax, k, n, s):
    f_max = 0
    alpha_max = None
    alpha = random_alpha(size=k*n)
    T = []
    for time in range(tmax):
        f_loc_max = 0
        alpha_loc_max = None
        t_denied = None
        for t in range(len(alpha)):
            alpha_transformed = alpha[:t] + str(1 + int(alpha[t])*-1) + alpha[t+1:]
            f_x = f(alpha_transformed)
            if t not in T and f_x > f_loc_max:
                f_loc_max =  f_x
                alpha_loc_max = alpha_transformed
                t_denied = t
        alpha = alpha_loc_max
        if f_loc_max > f_max:
            f_max = f_loc_max
            alpha_max = alpha_loc_max
        if len(T) == s:
            T.pop(0)
        T.append(t_denied)

        print(f"t: {time}, alpha: {alpha}, f: {f_x}, alpha_max: {alpha_max}, f_max: {f_max}")
    
    return (alpha_max, f_max)
```

```python
print(random_search(lambda alpha: hodnota_batohu(alpha, a, M), tmax = 100, k = len(a), n = 1))
```

```python
print(hill_climb(lambda alpha: hodnota_batohu(alpha, a, M), tmax = 100, k = len(a), n = 1, c0=10, pmut=0.25))
```

```python
print(tabu_search(lambda alpha: hodnota_batohu(alpha, a, M), tmax = 100, k = len(a), n = 1, s=5))
```
